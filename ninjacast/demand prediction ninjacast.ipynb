### **merge data**

```python
fact.columns
```

```python
# convert order_date and date into same formate
fact["order_date"] = pd.to_datetime(fact["order_date"]).dt.date
date["date"] = pd.to_datetime(date["date"]).dt.date
```

```python
#merge the all datsets
df = fact.merge(product, on="product_id", how="left") \
         .merge(region, on="region_id", how="left") \
         .merge(supplier, on="supplier_id", how="left") \
         .merge(warehouse, on="warehouse_id", how="left") \
         .merge(date, left_on="order_date", right_on="date", how="left")
```

### **data cleaning**

```python
# removing duplicates
df=df.drop_duplicates()
```

```python
# chaking missing values
df.isnull().sum()
```

```python
# hndel missing values
df=df.dropna()
```

```python
# fix data types
df.info()
```

## **feature engineering**

```python
# convert date time into int format
df["order_date"] = pd.to_datetime(df["order_date"])
```

```python
# extract date features
df["year"] = df["order_date"].dt.year
df["month"] = df["order_date"].dt.month
df["day"] = df["order_date"].dt.day
df["weekday"] = df["order_date"].dt.weekday
```

```python
# selecting target columns
y=df["quantity_ordered"]  #for demand
```

```python
# for x
x = df[["unit_price",
    "transport_cost",
    "operational_cost",
    "delay_days",
    "year",
    "month",
    "weekday"
]]
```

```python
print("x :",x.head())
print("y :",y.head())
```

### **train test split**

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42
)
```

## **train model**

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(random_state=42)
model.fit(x_train, y_train)
```

```python
# prediction
y_pred = model.predict(x_test)
```

### **evaliuate model**

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.metrics import r2_score

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2=r2_score(y_test,y_pred)

print("MAE:", mae)
print("MSE:", mse)
print("R2 score:",r2)
```

## **creating predict dataset**

```python
future_df=df.copy()
```

```python
# convert date
future_df["order_date"]=pd.to_datetime(future_df["order_date"])
```

```python
# shift date to future date
future_df["order_date"] = future_df["order_date"].apply(lambda x: x.replace(year=2026))
```

```python
# creating date features
future_df["year"] = future_df["order_date"].dt.year
future_df["month"] = future_df["order_date"].dt.month
future_df["weekday"] = future_df["order_date"].dt.weekday
```

```python
# adding quantity order
future_df["quantity_ordered"] = future_df["quantity_ordered"] * 1.05
future_df["unit_price"] = future_df["unit_price"] * 1.03
```

```python
# prepare features
future_features=future_df.reindex(columns=x.columns)

# fill missing values
future_features=future_features.fillna(0)
```

```python
# predict future demand
future_df["predicted_demand"] = model.predict(future_features)
```

```python
# selecting only important columns for dataset
final_df = future_df[[
    "product_id",
    "product_name",
    "region_id",
    "region_name",
    "warehouse_id",
    "warehouse_name",
    "order_date",
    "quantity_ordered",
    "quantity_supplied",
    "predicted_demand"
]]
```

```python
# rename the columns
final_df = final_df.rename(columns={
    "order_date": "predict_Date",
    "quantity_ordered": "Actual_Demand",
    "quantity_supplied": "Supply",
    "predicted_demand": "Predicted_Demand"
})
```

```python
# remove the duplicate
final_df=final_df.drop_duplicates()
```
